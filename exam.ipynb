Импортируем необходимые модули:

```python
from bs4 import BeautifulSoup
import requests, time, re
```

Пишем функцию, принимающую на вход словарь, значение которого ссылки (изначально это словарь из одной исходной ссылки), словарь, ключи которого термин, а значение - ссылка на страницу википедии про этот термин, счётчик и пороговое значение переходов по ссылкам. 
Наша функция проходится по ссылкам википедии (печатает саму ссылку, чтобы можно было следить за процессом). Функция вытаскивает ссылки и отбрасывает ненужные. Потом из всех нужных ссылок с помощью регулярных выражений она достаёт название тремина и ссылку на него, записывает это в формате словаря. После всего этого к каунтеру прибавляется единица. Если в этот момент каунтер доходит до порога, функция просто вернёт термины, если нет - запустит сама себя для всех имеющихся статей.

```python
def get_all_s(wikilist, terms, counter, porog):
    termings = {}
    for wiki in wikilist.values():
        print(wiki)
        lev = []
        time.sleep(2)
        response = requests.get(wiki)
        soup = BeautifulSoup(response.text)
        ns = soup.find_all('div')
        for n in ns:
            if 'mw-content-text' in str(n):
                a = n.find_all('a')
                for ssyl in a:
                    if "title" not in str(ssyl):
                        pass
                    elif 'Категория' in str(ssyl):  #категории по идее не термин, а ссылка на список страниц, так что я их убираю
                        pass
                    elif 'Редактировать' in str(ssyl):
                        pass
                    elif 'extiw' in str(ssyl):
                        pass
                    elif 'en.' in str(ssyl):
                        pass
                    else:
                        lev.append(str(ssyl))

        for el in lev:
            pattern = 'title="[А-я\d\s.,()A-z—»«-]+"'
            p = re.search(pattern, el)
            if p == None:
                pass
            elif 'страница отсутствует' in p[0]:
                pass
            else:
                patt = 'href="[A-z%\d_./]+'
                s = re.search(patt, el)
                if s == None:
                    pass
                else:
                    pew = p[0].replace('title="', '')
                    pew = pew.replace('"', '')
                    se = s[0].replace('href="', '')
                    sew = 'https://ru.wikipedia.org' + se
                    new = {pew: sew}
                    termings.update(new)
    terms.update(termings)
    counter = counter + 1
    if counter < porog:
        a = get_all_s(terms, terms, counter, porog)
    return terms
```

Создаём вводные данные и запускаем функцию:

```python
wl = {}
terms = {}
page = input('Введите страницу вики: ')
name = input('Введите её название: ')
pn = {name: page}
wl.update(pn)

porog = int(input('Сколько раз переходить по ссылкам? Введите число не меньше 1: '))
termins = get_all_s(wl, terms, 0, porog)
ters = list(set(termins))
```
